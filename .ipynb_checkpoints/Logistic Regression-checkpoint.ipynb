```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
import matplotlib.pyplot as plt
plt.rc("font", size = 14)
from sklearn.linear_model import LogisticRegression
# from sklearn.cross_validation import train_test_split
from sklearn.model_selection import train_test_split
import seaborn as sns
sns.set(style = "white")
sns.set(style = "whitegrid", color_codes = True)
```

```python
data.drop(data.columns[[0,3,7,8,9,10,11,12,13,15,16,17,18,19]], axis = 1, inplace = True)
```

```python
##Data Preprocessing
data2 = pd.get_dummies(data, columns = ['job','marital','default','housing', 'loan', 'poutcome'])
```

```python
# Drop unknown columns
data2.drop(data2.columns[[12,16,18,21,24]], axis = 1, inplace = True)
data2.columns
```

```python
# Check independence between the independent variables
sns.heatmap(data2.corr())
plt.show()
```

```python
# Splitthe data into training and test sets
x = data2.iloc[:,1:]
y = data2.iloc[:,0]
x_train, x_test, y_train, y_test = train_test_split(x,y, random_state = 0)
```

```python
# Check out training data is sufficient
x_train.shape
```

```python
## Logostic Regression Model
# Fit logistic regression to training set
classifier = LogisticRegression(random_state = 0)
classifier.fit(x_train,y_train)
```

```python
# Predicing the test set and creating confusion matrix
y_pred = classifier.predict(x_test)
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(y_test, y_pred)
print(confusion_matrix)
```

```python
# Accuracy
print('Accuracy of logistic regression classifier on test set: {:.2f}'.format(classifier.score(x_test, y_test)))
```

```python
# Compute precision, recall, F-measure and support
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
